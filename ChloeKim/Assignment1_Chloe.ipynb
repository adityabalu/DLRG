```python
# 9-12

class Robot():
    
    def __init__(self, position, direction, time, speed):
        
        self.position = position
        self.direction = direction
        self.time = time
        self.speed = speed
        
    def __repr__(self):
        return ('Current position is {0}, current direction is {1}, current time is {2}.' .format(self.position, self.direction, self.time))

    def wait(self, time): 
        self.time += time
        return self.time
    
    def move_forward(self, time): 
        self.position[0] += self.speed*time*self.direction[0] # current direction에서 time만큼 간 거리 더하기
        self.position[1] += self.speed*time*self.direction[1]
        self.time += time
        
    def rotate_right(self):      
        x = 0*self.direction[0] - (-1)*self.direction[1]
        y = 1*self.direction[0]+ 0*self.direction[1]
        self.direction = [x, y]

    def rotate_left(self): 
        x = 0*self.direction[0] - (1)*self.direction[1]
        y = (-1)*self.direction[0]+ 0*self.direction[1]
        self.direction = [x, y]

def move_square(robot):
    time = 10/0.5
    print(robot)
    
    robot.move_forward(time) # [side_length, 0]
    print(robot)
    robot.wait(5)
    print(robot)
    robot.rotate_left()
    print(robot)
    robot.move_forward(time) # [side_length, side_length]
    print(robot)
    robot.wait(5)
    print(robot)
    robot.rotate_left()
    print(robot)
    robot.move_forward(time) # [0, side_length]
    print(robot)
    robot.wait(5)
    print(robot)
    robot.rotate_left()
    print(robot)
    robot.move_forward(time) # [0, 0]
    print(robot)
    robot.wait(5)
    print(robot)

cur_position = [0,0]
cur_direction = [0,1]
cur_time = 0
cur_speed = 0.5    

robot = Robot(cur_position, cur_direction, cur_time, cur_speed)
move_square(robot)
```
