```python
Distances  =  np.array([0,  198,  303,  736,  871,  1175,  1475, 1544, 1913, 2448]) 

distance_matrix = np.zeros((len(Distances), len(Distances)))
#print(distance_matrix)

for i in range(len(Distances)):
    for j in range(len(Distances)):
        if i == j:
            distance_matrix[i,j] = 0
        elif i<j:
            distance_matrix[i,j] = Distances[j] - Distances[i]
        elif i>j:
            distance_matrix[i,j] = Distances[i] - Distances[j]
# print("Distance matrix : ")
# print(distance_matrix)

city_names = ["A", "B", "C", "D", "E", "F", 
              "G", "H", "I", "J"]

fx, ax = plt.subplots()
im = ax.imshow(distance_matrix, cmap="viridis")

# Label each axis with a city name 
ax.set_xticks(np.arange(len(Distances)))
ax.set_yticks(np.arange(len(Distances)))
ax.set_xticklabels(city_names)
ax.set_yticklabels(city_names)

# Add a colorbar 
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel("Distance ", rotation=-90, va="bottom")
# Show the plot 
plt.show()
```

```python
folderName = ''

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

def createFolderAutoName(directory):
    now = datetime.datetime.now()
    folderName = now.strftime("%Y%m%d-%H%M%S") # folder name
    pathName = os.path.join(directory, folderName)
  
    createFolder(pathName)
    # Save the changed parameters in the folder
    with open('./run/' + folderName +'/vegetables.yaml', 'w') as f:
        yaml.dump(vegetables, f)    
    
# Create the run folder
createFolderAutoName('./run')
```
