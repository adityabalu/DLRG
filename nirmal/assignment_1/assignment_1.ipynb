```python
'''section 1 problem 1'''
def check_strobogrammatic(number)->bool:
    '''this function takes a number and checks if it is strobogrammatic'''
    assert (number%1 == 0), "please gimme an integer"
    
    # strobogrammatic digits
    original = [0,1,6,9,8]
    mirror = [0,1,9,6,8]

    check = False
    count = 0  

    number = str(number)
    # if odd half_length = (num+1)/2, if even half_length = (num)/2
    half_length = int(ceil(len(number)/2)) 

    for i in range(half_length):
        for j in range(len(original)):
            if (int(number[i])==original[j] and int(number[len(number)-(i+1)])==mirror[j]):
                count +=1 #counting if any digit satisfy strobogrammatic criteria

    if count == half_length: # checking if all digits satisfy strobogrammatic criteria
        check = True

    return check

def strobogrammatic(num_digit=1)->list:
    '''takes number of digits and spit out all the strobogrammatic numbers'''
    assert (num_digit%1 == 0), "Number of digit can only be an integer"
    lst = []
    for i in itertools.product(['0','1','6','8','9'], repeat=num_digit):
        number = int(''.join(i))
        if len(str(number))==num_digit:  # dropping numbers with leading zeros
            if check_strobogrammatic(number):
                lst.append(str(number))
    return lst

# test results
print(strobogrammatic(1)) 
print(strobogrammatic(2))
print(strobogrammatic(3))
print(strobogrammatic(4))
```

```python
'''section 1 problem 2'''
def string_lower(word):
    assert isinstance(word, str), 'input must be string'
    lower_case_word = ""
    for s in word:
        lower_case_word+=s.lower()

    return lower_case_word
     
def string_concat(lst1, lst2):
    assert isinstance(lst1[0], str), 'first element of the first list must be a str.'
    assert isinstance(lst1[1], str), 'second element of the first list must be a str.'
    assert isinstance(lst2[0], str), 'first element of the second list must be a str.'
    assert isinstance(lst2[1], str), 'second element of the second list must be a str.'
    
    for i in range(len(lst1)):
        lst1[i] = string_lower(lst1[i])

    for i in range(len(lst2)):
        lst2[i] = string_lower(lst2[i])
    
    zipped = zip(lst1, lst2)

    lst = [x[0]+' '+x[1] for x in list(zipped)]

    return lst

lst1=['hello', 'Bye']
lst2=['WORLD', 'wOrld']
string_concat(lst1, lst2)
```

```python
'''section 1 problem 3'''
def product_list(lst1: list, lst2: list)->list:
    assert isinstance(lst1, list), 'first parameter must be a list'
    assert isinstance(lst2, list), 'second parameter must be a list'

    zipped = list(zip(lst1, lst2))
    # the following line checks if both element of the zipped list is a numeric quantity
    new_zipped = [x for x in zipped if (str(x[0]).isdigit() and str(x[1]).isdigit())]

    product =  [float(x[0])*float(x[1]) for x in new_zipped]

    return product

lst1 = ['12', '13', 'Hello', '4', '6', '10]']
lst2 = ['2', '3', '4', 5, '[6',  12, 14]
product_list(lst1, lst2)
```

```python
'''section 2 problem 9,10,11,12'''
import numpy as np
import time as tm
class Robot:
    def __init__(self):
        self.position = np.zeros((2,1))
        self.direction = np.array([[1],[0]])
        self.time = 0.0
        self.speed = 0.5

    def __repr__(self):
        return f'Robot(position = {self.position.tolist()}, direction = {self.direction.tolist()}, time = {self.time}, speed = {self.speed})'
    
    def wait(self,time):
        self.time += time
    
    def move_forward(self, time):
        self.position += self.direction * self.speed * time
        self.time += time
    
    def rotate_right(self):
        right_rotation_matrix = np.array([[0,1],[-1,0]])
        self.direction = right_rotation_matrix@self.direction

    def rotate_left(self):
        left_rotation_matrix = np.array([[0,1],[-1,0]])
        self.direction = left_rotation_matrix@self.direction

    def move_square(self, side_length):
        print('Initial state of the robot is')
        print(self.__repr__())
        print('\n')
        for k in range(4):
            self.move_forward(side_length)
            print('Robot finished moving. Current state is:')
            print(self.__repr__())
            print('\n')
            self.rotate_left()
            print('Robot rotated left. Current state is:')
            print(self.__repr__())
            print('\n')
        
        print('final state of the robot is')
        print(self.__repr__())
        
myRobot = Robot()
myRobot.move_square(5)
```
