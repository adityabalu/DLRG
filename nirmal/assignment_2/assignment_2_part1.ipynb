```python
'''
Q4. Given an array of distances between 10 cities, calculate the distance of each city 
from every other city and plot it using matplotlib using a colorbar, label the axes, 
include city names. E.g. Distances = np.array([0, 198, 303, 736, 871, 1175, 1475, 
1544, 1913, 2448])
'''


import numpy as np
import matplotlib.pyplot as plt

distances = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])
distance_matrix = np.abs(distances - distances[:, np.newaxis])


fig, ax = plt.subplots(figsize=(10,10))

im = ax.imshow(distance_matrix, cmap='plasma')
cbar = ax.figure.colorbar(im, ax=ax)

# set the tick labels for the x and y axes
ax.set_xticks(np.arange(len(distances)))
ax.set_yticks(np.arange(len(distances)))
ax.set_xticklabels(['City{}'.format(i+1) for i in range(len(distances))])
ax.set_yticklabels(['City{}'.format(i+1) for i in range(len(distances))])

# set the axis labels and title
ax.set_xlabel('Destination Cities')
ax.set_ylabel('Starting Cities')
ax.set_title('City Distance Matrix')

# loop over data dimensions and create text annotations
for i in range(len(distances)):
    for j in range(len(distances)):
        text = ax.text(j, i, distance_matrix[i, j],
                       ha="center", va="center", color="w")

# show the plot
plt.show()
```

```python
'''
Q5. 
Assume T is a multidimensional list. Suppose we make another variable Q equal 
to T and change an element in T. You need to print out Q and T and check what 
happens. Do they both have the same elements? If so, why? What can we do to Q 
to not be affected by operations on T? (Hint: deep copy)
e.g.:
T = [[1, 2, 3], [ 4, 5, 6], [7, 8, 9]]
Q = list(T)
T[1][0] = 100000
'''

import copy

T = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
Q = list(T)
T[1][0] = 100000

print("T =", T)
print("Q =", Q)

Q_deepcopy = copy.deepcopy(T)
T[2][2] = 9999

print("T =", T)
print("Q_deepcopy =", Q_deepcopy)


# simpler example
x = [1,2,3]
y = x.copy()
x[0] = 100
print(x)
print(y)
```

```python
'''
Q6
For given two arrays, find
a. The indices as well as the elements in X where the current value is greater 
than the corresponding value in Y
b. The indices of elements in X that are equal to elements in Y
X = np,array([5, 9, 77, 0, 74, 55, 81, 79, 90])
Y = np.array([5, 1, 46, 2, 10, 29, 55, 4, 90])
'''

import numpy as np

X = np.array([5, 9, 77, 0, 74, 55, 81, 79, 90])
Y = np.array([5, 1, 46, 2, 10, 29, 55, 4, 90])

greater_indices = np.where(X > Y)[0]
greater_elements = X[greater_indices]

print(f"Indices of elements in X where the current value is greater than the corresponding value in Y: {greater_indices}")
print(f"Elements in X where the current value is greater than the corresponding value in Y:{greater_elements}")

equal_indices = np.where(X == Y)[0]
equal_elements = X[equal_indices]

print(f"Indices of elements in X where the current value is equal to the corresponding value in Y: {equal_indices}")
print(f"Elements in X where the current value is greater than the corresponding value in Y: {equal_elements}")
```
