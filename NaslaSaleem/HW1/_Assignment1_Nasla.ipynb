```python
#Q1
import re
strobo_numbers=['0','1','8']
def strobo_check(number):
  flag=1
  for i in range(0,len(number)//2):
    if((number[i]==number[-i-1]) and (number[i] in strobo_numbers)):
      continue
    elif((number[i]=='6' and number[-i-1] =='9') or (number[i]=='9' and number[-i-1] =='6')):
      continue
    else: 
      flag=0
  return flag
def strobo_even(number):
      tmp_mirror=re.match('(69)(69)*',number)
      stem=0
      if(tmp_mirror!=None):
        span=tmp_mirror.span()
        if(span[1]-span[0]==len(number)):
          stem=1
          return number
      if(stem==0):
        flag=strobo_check(number)
        if(flag==1):
          return number
        else:
          return False

def strobo_function(number):
  tmp=re.search('0+|1+|8+',number)
  if(tmp!=None):
        span=tmp.span()
        if(span[1]-span[0]==len(number)):
          return number
  if(len(number)==1 and number in strobo_numbers):
    return number
  elif(len(number)%2==0):  
    if(strobo_even(number)):
      return number
    else:
      return False    
  else:
    middle_ix=len(number)//2
    if(number[middle_ix] in strobo_numbers):
      left=number[:middle_ix]
      right=number[middle_ix+1:]
      result=strobo_even(left+right)
      
      if(result):
          return number
      else:
          return False
    else:
        return False
```

```python
number=input('Enter the number: ')
strobo_list=[]
if(int(number)==1):
  start_size=0
else:
  start_size=10**(int(number)-1)
for i in range(start_size,10**int(number)):
  string=str(i)
  result=strobo_function(string)
  if(result):
    strobo_list.append(result)
print('List for n='+number)
print(strobo_list)
print(len(strobo_list))
```

```python
#Q7
import time
import math
import numpy as np
import random
# decorator to calculate duration
# taken by any function.
def calculate_time(func):   
    # added arguments inside the inner1,
    # if function takes any arguments,
    # can be added like this.
    def inner1(*args, **kwargs):
 
        # storing time before function execution
        begin = time.time()
         
        func(*args, **kwargs)
 
        # storing time after function execution
        end = time.time()
        print("Total time taken in : ", func.__name__, end - begin)
 
    return inner1


#Q8
# this can be added to any function present,
# in this case to calculate a factorial

#a)
@calculate_time
def pdt_zip(l1,l2):
  result=[]
  for i in zip(l1,l2):
    result.append(i[0]*i[1])
  print('Product of  lists using zip: '+str(result))
  return result
  
@calculate_time
def pdt_ind(l1,l2):
  result=[]
  for i in range(0,len(l1)):
    result.append(l1[i]*l2[i])
  print('Product of  lists accessing individually: '+str(result))
  return result

@calculate_time
def pdt_listcomprehension(l1,l2):
  result= [l1[i]*l2[i] for i in range(len(l1))]
  print('Product of  lists using list comprehension: '+str(result))
  return result

#c)
@calculate_time
def uniform_distribution_list():
  result=list(np.random.uniform(0, 1000, 10000))
  print(result)
  return result

#d)
@calculate_time
def uniform_distribution_tuple():
  result=tuple(np.random.uniform(0, 1000, 10000))
  print(result)
  return result

#e)
@calculate_time
def tuple_median():
  distribution=tuple(np.random.uniform(0, 1000, 10000))
  distribution=sorted(distribution)
  if(len(distribution)%2!=0):
    median=distribution[len(distribution)//2 -1]
  else:
    median=(distribution[len(distribution)//2 -1]+distribution[len(distribution)//2])/2
  print('Median:' +str(median))
  return median

@calculate_time
def list_median():
  distribution=list(np.random.uniform(0, 1000, 10000))
  distribution=sorted(distribution)
  if(len(distribution)%2!=0):
    median=distribution[len(distribution)//2 -1]
  else:
    median=(distribution[len(distribution)//2 -1]+distribution[len(distribution)//2])/2
  print('Median:' +str(median))
  return median

#f) List. Becuase the elements of a list are modifiable whereas for a tuple, it's immutable
#g)
@calculate_time
def distributions():
  distr={}
  mu_gauss = 100
  sigma_gauss =50  
  mu_lgn = 0
  sigma_lgn = 0.25
  mu_nv=100
  sigma_nv=50
  distr['gauss']=[]
  distr['lognormvariate']=[]
  distr['normalvariate']=[]
  for i in range(0,1000):
      distr['gauss'].append(random.gauss(mu_gauss, sigma_gauss)) 
      distr['lognormvariate'].append(random.lognormvariate(mu_lgn, sigma_lgn)) 
      distr['normalvariate'].append(random.gauss(mu_nv, sigma_nv)) 
  print(distr)
  return distr
```

```python
class Robot:
  def __init__(self, position=[0,0], direction=[1,0], set_time=0.0, speed = 0.5):

    self.position = [0,0] # Origin
    self.possible_directions=[[1,0],[0,-1],[-1,0],[0,1]]
    self.direction = direction # (1,0) -> right, # (-1,0) -> left, # (0,1)-> forward
    self.time = set_time
    self.speed = speed # Lets consider this 0.5m/s
    self.num = 0

  def __repr__(self) -> str: 
    #return self.position, self.direction, self.time, self.speed
    return ('Robot position is {}, direction is {} and time elapsed is {}'.format(self.position,self.direction,self.time))
  
  def wait(self, set_time):
    start_time = time.time()
    self.speed = 0.0
    while((time.time()- start_time) < set_time):
      continue
    self.time = self.time + set_time
  def move_forward(self, set_time, speed): 
    start_time = time.time()
    self.speed = speed
    while((time.time()- start_time) < set_time):
      # print('Time elapsed is'+ str(time.time()-start_time))
      # if (time.time()-start_time > 5):
      #   break
      continue
    distance = self.speed * set_time
    self.time = self.time + set_time
    if self.direction == [-1,0]:
      self.position[0] = self.position[0] - distance
    elif self.direction == [1,0]:
      self.position[0] = self.position[0] + distance
    elif self.direction == [0,1]:
      self.position[1] = self.position[1] + distance
    elif self.direction == [0,-1]:
      self.position[1] = self.position[1] - distance
    
  def rotate_right(self):
    # Stop the robot and turn right
    self.speed = 0.0
    self.num= self.num +1
    self.direction = self.possible_directions[self.num]

  def rotate_left(self): 
    # Stop the robot and turn left
    self.speed = 0.0
    self.num = self.num -1
    self.direction = self.possible_directions[self.num]
    print('Direction is {}'.format(self.direction))
```

```python
def move_square(side_length):
  speed = 10
  mycar = Robot(speed=speed)
  print(repr(mycar))
  set_time = side_length/mycar.speed
  print('moving the robot for {} seconds'.format(set_time))
  mycar.move_forward(set_time, speed=speed)
  print(repr(mycar))
  mycar.rotate_left()
  print(repr(mycar))
  mycar.move_forward(set_time, speed=speed)
  print(repr(mycar))
  mycar.rotate_left()
  print(repr(mycar))
  mycar.move_forward(set_time, speed=speed)
  print(repr(mycar))
  mycar.rotate_left()
  print(repr(mycar))
  mycar.move_forward(set_time, speed=speed)
  print(repr(mycar))

move_square(10)
```
