```python
# question 1
def decorator_unique(func):
    def wrapper(*args):
        return np.unique(np.array(func(*args)))
    return wrapper

@decorator_unique
def strobogrammatic(n):

    def helper(input_list,n):

        new_items =[]
        if n == 2 :
            for item in input_list:
                new_items.append("9"+item+"6")
                new_items.append("6"+item+"9")
                new_items.append("8"+item+"8")
                new_items.append("1"+item+"1")
        if n == 1:
            for item in input_list:
                item_len_mid = len(item) // 2
                if len(item) % 2 == 0:
                    new_items.append(item[:item_len_mid]+ '0' + item[item_len_mid:])
                    new_items.append(item[:item_len_mid]+ '1' + item[item_len_mid:])
                    new_items.append(item[:item_len_mid]+ '8' + item[item_len_mid:])

        return new_items

    if n==0:
        return ['']

    if n==1:
        return ['0','1','8']

    if n >= 2 :
        return [*helper(strobogrammatic(n-2),2), *helper(strobogrammatic(n-1),1)]
```

```python
# test question 1
strobogrammatic(3)
```

```python
# question 2
from typing import List

def concatenate_same_length_list_lowercase(list_1: List[str] , list_2: List[str] ) -> List[str]:
    result_list = [x + " " + y for x,y in zip(list_1, list_2)]
    return [x.lower() for x in result_list]
```

```python
# test question 2
list1 = ["HellO", "BYe"]
list2 = ["woRlD", "WOrld"]

concatenate_same_length_list_lowercase(list1,list2)
```

```python
# question 3
def multiply_num_list(list_1, list_2):
    min_size = min(len(list_1),len(list_2))
    def helper(x):
        if type(x)== str:
            if x.isnumeric():
                return True
        else:
            return False

    return [int(x) * int(y) for x,y in zip(list_1[:min_size],list_2[:min_size]) if helper(x) and helper(y)]

```

```python
# test question 3

list1 = ["12", "13", "Hello", "4", "6", "10"]
list2 = ["2", "3", "4", "5", ["6", "12", "14"]]
multiply_num_list(list1,list2)
```

```python
# question 7
import time
def decorator_timeit(func):
    def wrapper_func(*args, **kwargs):
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()
        elapsed_time = end_time - start_time
        print(f'Function {func.__name__} take {elapsed_time} secend to run')
        return result
    return wrapper_func

@decorator_timeit
def sleep_function(n):
    time.sleep(n)
    pass
```

```python
import time
import numpy as np

class Robot:
    """
    this class describe the situation of a robot
    """
    def __init__(self, current_position=None, current_direction=None, current_time=0.0, current_speed=0.5):
        if current_direction is None:
            current_direction = np.array([1, 0])
        if current_position is None:
            current_position = np.array([0, 0])

        self.current_position = current_position
        self.current_direction = current_direction
        self.current_time = current_time
        self.current_speed = current_speed


    def __repr__(self):
        return f'current position is: {self.current_position} \ncurrent direction is: {self.current_direction} \ncurrent time is: {self.current_time}'

    def wait(self,time_sec):

        time.sleep(time_sec)

    def move_forward(self,time_sec):
        passed_trajectory = self.current_direction * self.current_speed * time_sec
        self.current_position  =  np.add(self.current_position,passed_trajectory)
        self.current_time += time_sec

    def rotate_right(self):
        rotation_matrix = np.array([[0,1],[-1,0]])
        self.current_direction = np.dot(rotation_matrix,self.current_direction)

    def rotate_left(self):
        rotation_matrix = np.array([[0,-1],[1,0]])
        self.current_direction = np.dot(rotation_matrix,self.current_direction)

    def move_square(self,side_length=10):
        time_per_side =  side_length /  self.current_speed
        # first side
        self.move_forward(time_per_side)
        print(self.__repr__())
        self.rotate_left()
        # second side
        self.move_forward(time_per_side)
        print(self.__repr__())
        self.rotate_left()
        # third side
        self.move_forward(time_per_side)
        print(self.__repr__())
        self.rotate_left()
        # fourth side
        self.move_forward(time_per_side)
        print(self.__repr__())




```

```python
ro = Robot()
```
